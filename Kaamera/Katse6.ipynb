```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# --- 1. Pildi laadimine ---
path = '/content/drive/MyDrive/Volans_camera/test16.jpg'
pilt = cv2.imread(path)
if pilt is None:
    raise ValueError("Pilti ei leitud – kontrolli failinime ja asukohta")

# --- 2. Kanalid (NRG pilt) ---
NIR = pilt[:, :, 2].astype(float)  # R kanal = NIR
R = pilt[:, :, 1].astype(float)    # G kanal = Red

# --- 3. NDVI ---
ndvi = (NIR - R) / (NIR + R + 1e-10)

# --- 4. KUVA TULEMUS (ei salvesta kuhugi) ---
plt.figure(figsize=(6,5))
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title("ortofoto")
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import time
import gspread

# Path to your CSV
file_path = 'data_csv'

# Google Sheet link
sheet_url = "https://docs.google.com/spreadsheets/d/1zRlm4kUXhSytt9hO08TRbgX_2O0CUImVtxnR7uCV9x4/edit?usp=drivesdk"

# Extract sheet ID
sheet_id = sheet_url.split("/d/")[1].split("/")[0]

# Connect to the Google Sheet
gc = gspread.service_account(filename=None) # None because public edit
sh = gc.open_by_key(sheet_id)
worksheet = sh.sheet1

# Keep track of how many lines we have already sent
last_line_count = 0

while True:
    df = pd.read_csv(file_path)

    if len(df) > last_line_count:
        new_rows = df.iloc[last_line_count:]
        last_line_count = len(df)

        for _, row in new_rows.iterrows():
            worksheet.append_row(row.tolist())
            print("Sent to Google Sheets:", row.tolist())

    time.sleep(2) # check every 2 seconds
```

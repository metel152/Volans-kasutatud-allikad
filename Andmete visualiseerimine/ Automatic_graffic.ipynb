```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Google Sheet setup
# ----------------------------
sheet_url = "https://docs.google.com/spreadsheets/d/1zRlm4kUXhSytt9hO08TRbgX_2O0CUImVtxnR7uCV9x4/edit?usp=drivesdk"

# Convert to CSV export URL
csv_url = sheet_url.replace("/edit?usp=drivesdk", "/gviz/tq?tqx=out:csv")

# ----------------------------
# Step 2: Read latest data
# ----------------------------
df = pd.read_csv(csv_url)

# Convert timestamp to datetime for proper plotting
df['timestamp'] = pd.to_datetime(df['timestamp'])

# ----------------------------
# Step 3: Plotting all sensors
# ----------------------------
plt.figure(figsize=(14,6))

plt.plot(df['timestamp'], df['temperatue_C'], label='Temperature (°C)', marker='o')
plt.plot(df['timestamp'], df['pressure_hPa'], label='Pressure (hPa)', marker='x')
plt.plot(df['timestamp'], df['gps_altitude_m'], label='Altitude (m)', marker='s')

plt.xlabel('Timestamp')
plt.ylabel('Sensor Values')
plt.title('LoRa Sensor Data from Google Sheets')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------------
# Step 4: Optional separate graph for Altitude
# ----------------------------
plt.figure(figsize=(12,4))
plt.plot(df['timestamp'], df['gps_altitude_m'], color='orange', label='Altitude (m)')
plt.xlabel('Timestamp')
plt.ylabel('Altitude (m)')
plt.title('GPS Altitude Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your combined CSV file
data = pd.read_csv("combined_data.csv")

# Convert timestamp column to datetime if you want to use it
data["timestamp"] = pd.to_datetime(data["timestamp"])

# --- Graph 1: Temperature vs GPS Altitude ---
plt.figure(figsize=(6, 8))
plt.plot(data["temperature_C"], data["gps_altitude_m"], color="red", marker="o", linestyle="-")
plt.xlabel("Temperature (°C)")
plt.ylabel("Altitude (m)")
plt.title("Temperature vs Altitude")
plt.grid(True)
plt.show()

# --- Graph 2: Pressure vs GPS Altitude ---
plt.figure(figsize=(6, 8))
plt.plot(data["pressure_hPa"], data["gps_altitude_m"], color="blue", marker="o", linestyle="-")
plt.xlabel("Pressure (hPa)")
plt.ylabel("Altitude (m)")
plt.title("Pressure vs Altitude")
plt.gca().invert_xaxis()  # Pressure decreases with altitude
plt.grid(True)
plt.show()
```

```python
csv_data = """timestamp,temperature_C,pressure_hPa,gps_altitude_m
2026-03-15 10:00:00,22.0,1013.2,10
2026-03-15 10:01:00,22.1,1013.0,12
2026-03-15 10:02:00,22.3,1012.8,14"""

with open('combined_data.csv', 'w') as f:
    f.write(csv_data)

print('combined_data.csv has been created.')
```

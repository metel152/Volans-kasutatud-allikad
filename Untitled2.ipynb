```python
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import cv2
import matplotlib.pyplot as plt

# --- 1. Pildi laadimine ---
path = '/content/drive/MyDrive/Volans_camera/test18.jpg'
pilt = cv2.imread(path)

if pilt is None:
    raise ValueError("Pilti ei leitud – kontrolli failinime ja asukohta")

# --- 2. Kanalid ---
NIR = pilt[:, :, 0].astype(float)
R = pilt[:, :, 2].astype(float)

# --- 3. NDVI ---
ndvi = (NIR - R) / (NIR + R + 1e-10)

# --- 4. Salvesta ---
np.save('/content/drive/MyDrive/Volans_camera/ndvi_andmed.npy', ndvi)

# --- 5. Näita pilti ---
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title("NDVI")
plt.axis('off')
plt.show()
```
